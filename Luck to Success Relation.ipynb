```python
# Each data will represent a person with 0 = skill, 1 = luck, z = combined score


import numpy as np
import pandas as pd
import random
```

```python
def new_population(total):
    luck = []
    skill = []
    for n in range(total):
        x=random.random()
        y=random.random()
        luck.append(y)
        skill.append(x)
    return pd.DataFrame(list(zip(skill,luck))) 
        
def sumscore(pop, luckpct):
    #assuming luck is a percentage of the success attribute or zscore
    pop['S'] = pop[0]*(1-luckpct)
    pop['L'] = pop[1]*luckpct
    pop['Z'] = pop['S'] + pop['L']
    return pop

def prodscore(pop):
    #assuming luck is a parameter that amplify the skills score
    pop['Z'] = (pop[0]**2) * (pop[1])
    pop['Z'].squeeze()
    return pop
```

```python
def select(pdata, top):
    pdata.sort_values(by = 'Z', ascending = False, inplace = True)
    return pdata.head(top)
            
def qualify(pdata, zthold):
    df = pdata[pdata['Z']>zthold]
    return df
    
    
```

```python
popu = new_population(1000000)
pop1 = sumscore(popu, 0.05)
pop2 = prodscore(popu)
```

```python
pop1x = qualify(pop1,0.80)
pop2x = qualify(pop2,0.60)

pop1y = select(pop1, 1000)
pop2y = select(pop2, 1000)
```
